Вам дан набор данных с фактическими значениями целевой переменной и предсказанными значениями:


Фактические значения

actual_values = np.array([100, 150, 200, 250])


Предсказанные значения

predicted_values = np.array([90, 140, 180, 270])

Задача студентов состоит в расчете среднеквадратичной ошибки (MSE) и средней абсолютной ошибки (MAE) для данного набора данных без использования модели. Студенты должны использовать правильные значения фактической переменной и предсказанные значения для расчета функций потерь. Они также должны проанализировать результаты и объяснить, какие значения функции потерь указывают на лучшую точность предсказания.

```python
import numpy  as np
import matplotlib.pyplot as plt
```

```python
actual_values = np.array([100, 150, 200, 250])
```

```python
predicted_values = np.array([90, 140, 180, 270])
```

Расчет среднеквадратичной ошибки MSE:

```python
n = len(actual_values)
mse = np.mean((actual_values - predicted_values) ** 2)
print(f"Среднеквадратичноая ошибка MSE: {mse}") 
```

Расчет средней абсолютной ошибки MAE:

```python
mae = np.mean(abs(actual_values - predicted_values))
print(f"Средняя абсолютная ошибка MAE: {mae}")
```

Визуализация

```python
plt.scatter(actual_values, predicted_values)
plt.xlabel("Фактические значения")
plt.ylabel("Предсказанные значения")
plt.title("График рассеяния фактических и предсказанных значений")
plt.show();
```

MSE показывает среднее значение квадратов ошибок и чувствительна к выбровам. MAE показывает среднее значение абсолютных ошибок. Чем ниже значение ошибки, тем точнее предсказания модели. Если взять RMSE из значения нашейго MSE, то есть корень квадратный из 250, получим 15.8. Это значение одного уже одного порядка с MAE. Как правило, на практике берут MSE если есть необходимость придать значение большим ошибкам, если нет или для анализа больших значений берут MAE. То есть нужно рассматвать каждую ситуацию индивидуально.
